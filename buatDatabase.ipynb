```python
import pandas as pd
import sqlite3 as sq3
```

```python
conn = sq3.connect("database.db")
```

```python
df_new_kamusalay = pd.read_csv('new_kamusalay.csv', encoding='cp1252')
df_new_kamusalay.head()
```

```python
df_new_kamusalay.to_sql('newKamusAlay', conn, if_exists = 'replace', index = False)
df_new_kamusalay
```

```python
df_abusive = pd.read_csv('abusive.csv', encoding="cp1252")
df_abusive.head()
```

```python
df_abusive.to_sql('abusive', conn, if_exists = 'append', index = False)
df_abusive
```

```python
conn.commit()
conn.close()
```
